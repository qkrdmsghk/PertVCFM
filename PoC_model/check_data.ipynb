```python
adata = datamodule.test_adata
```

```python
import pandas as pd
import numpy as np

# 假设您的 AnnData 对象名为 adata
# --------------------------------------------------------
# 0. 准备数据：计算 Control 和 Perturbation 组的基因平均表达值
# --------------------------------------------------------

print("--- 步骤 0: 计算基因平均表达值 ---")

# 1. 获取表达值的密集数组
if hasattr(adata.X, 'toarray'):
    X_dense = adata.X.toarray()
else:
    X_dense = adata.X

obs_df = adata.obs

# 2. 随机选择一个扰动基因进行批量分析
perturbation_genes_all = obs_df[obs_df['control'] == 0]['gene_name'].unique()
np.random.seed(0) 
# 只选择一个基因进行对比，与您代码一致
selected_gene = np.random.choice(perturbation_genes_all, size=1, replace=False)[0] 
print(f"随机选择的扰动基因: {selected_gene}")

# 3. 初始化列表存储所有分布的 DataFrame

control_dfs = []
# --- 3.1 整体平均值 ---
# 整体对照：所有基因在所有对照细胞中的平均表达值
control_mean_overall = pd.DataFrame(X_dense[obs_df['control'] == 1, :].mean(axis=0), columns=['expression_value'])
control_mean_overall['plot_group'] = 'Overall Control' 
control_dfs.append(control_mean_overall)

perturbation_dfs = []
# 整体扰动：所有基因在选定扰动细胞中的平均表达值
perturbation_mask_overall = obs_df['gene_name'] == selected_gene
perturbation_mean_overall = pd.DataFrame(X_dense[perturbation_mask_overall, :].mean(axis=0), columns=['expression_value'])
perturbation_mean_overall['plot_group'] = f'Overall Perturbation: {selected_gene
}'
perturbation_dfs.append(perturbation_mean_overall)


# --- 3.2 按 Batch 分组平均值 ---
batch_labels = obs_df['batch'].unique().tolist()

for b in batch_labels[:50]:
    # 筛选当前 batch 的数据
    batch_mask = obs_df['batch'] == b
    X_batch = X_dense[batch_mask, :]
    obs_batch = obs_df[batch_mask] # 必须使用子集的 obs 进行索引

    # Batch 对照：所有基因在当前 batch 对照细胞中的平均表达值
    control_mean_batch = pd.DataFrame(X_batch[obs_batch['control'] == 1, :].mean(axis=0), columns=['expression_value'])
    control_mean_batch['plot_group'] = f'Control (Batch {b})'
    control_dfs.append(control_mean_batch)

    # Batch 扰动：所有基因在当前 batch 选定扰动细胞中的平均表达值
    perturbation_mask_batch = obs_batch['gene_name'] == selected_gene
    perturbation_mean_batch = pd.DataFrame(X_batch[perturbation_mask_batch, :].mean(axis=0), columns=['expression_value'])
    perturbation_mean_batch['plot_group'] = f'Perturbation (Batch {b})'
    perturbation_dfs.append(perturbation_mean_batch)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 假设您的数据计算部分已经成功运行，
# 并且 control_dfs, perturbation_dfs, selected_gene, batch_labels 已经定义。
# control_dfs[0] 和 perturbation_dfs[0] 是 Overall 数据。
# control_dfs[1:] 和 perturbation_dfs[1:] 是按批次划分的数据。

# --------------------------------------------------------
# 1. 准备数据和布局
# --------------------------------------------------------

# 总共需要绘制的图的数量：Overall (1) + 所有 Batch (len(batch_labels))
num_plots = len(control_dfs) # control_dfs 包含了 Overall 和所有 batches

N_COLS = 10 # 每行显示 10 个图
N_ROWS = (num_plots + N_COLS - 1) // N_COLS # 向上取整

fig, axes = plt.subplots(N_ROWS, N_COLS, figsize=(5 * N_COLS, 5 * N_ROWS), squeeze=False)
axes = axes.flatten() # 将 axes 展平，方便循环迭代

# --------------------------------------------------------
# 2. 循环绘制所有 Overall 和 Batch 的对比图
# --------------------------------------------------------

# 构造一个包含所有分组标签的列表，用于标题
plot_labels = [control_dfs[0]['plot_group'].iloc[0].replace('Overall Control', 'Overall')]
for b_label in batch_labels:
    plot_labels.append(f'Batch {b_label}')

for i in range(num_plots):
    ax = axes[i]
    
    # 获取当前 Control 和 Perturbation 的 DataFrame
    current_control_df = control_dfs[i]
    current_perturbation_df = perturbation_dfs[i]
    
    # 合并数据
    current_data = pd.concat([current_control_df, current_perturbation_df], ignore_index=True)
    
    # 获取标题标签
    plot_title = plot_labels[i]

    # 绘制 KDE
    sns.kdeplot(
        data=current_data,
        x='expression_value',
        hue='plot_group',
        fill=True,
        common_norm=False,
        palette=['grey', 'red'], # 保持 Control/Perturbation 的颜色一致
        ax=ax
    )

    # 设置标题和标签
    ax.set_title(f'{plot_title} Distribution', fontsize=12)
    ax.set_xlabel('Mean Gene Expression Value')
    ax.set_ylabel('Density')
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    ax.legend(title='', fontsize=8)

# --------------------------------------------------------
# 3. 清理和显示
# --------------------------------------------------------

# 隐藏多余的子图
for i in range(num_plots, N_ROWS * N_COLS):
    fig.delaxes(axes[i])

plt.suptitle(f'Gene Mean Expression Distributions (Perturbation: {selected_gene})', fontsize=18, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 0.98]) # 调整布局以适应 suptitle
plt.show()
```
